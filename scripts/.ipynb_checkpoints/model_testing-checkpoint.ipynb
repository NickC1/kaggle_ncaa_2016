```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
%matplotlib inline
sns.set_context('notebook',font_scale=1.5)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
def log_loss(prob,actual):
    LL = actual*np.log(prob) + (1-actual)*np.log(1-prob)
    LL = -np.mean(LL)
    return LL
```

```python
import sklearn.datasets as data
cmap='rainbow'
```

```python
X_data,y_data = data.make_blobs(n_samples=200,n_features=2,centers=2,
                      random_state=2,cluster_std=15)
```

```python
plt.scatter(X_data[:,0],X_data[:,1],c=y_data,cmap=cmap);
```

```python
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        X_data, y_data, test_size=0.33, random_state=42)
# create the regression object
logregr = linear_model.LogisticRegression()

# train the model
logregr.fit(X_train, y_train)
logregr.score(X_test,y_test)
```

```python
proba = logregr.predict_proba(X_test)
```

```python
log_loss(proba[:,1],y_test)
```

```python
import skflow
```

```python
from sklearn import datasets, metrics, cross_validation
```

```python
from sklearn import preprocessing
xd = preprocessing.normalize(X_data,axis=0)
```

```python
plt.scatter(xd[:,0],xd[:,1])
```

```python
#iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(xd, y_data,
    test_size=0.2, random_state=42)

# Build 3 layer DNN with 10, 20, 10 units respecitvely.
classifier = skflow.TensorFlowDNNClassifier(hidden_units=[5, 4, 5],
    n_classes=2, steps=100000)

# Fit and predict.
classifier.fit(X_train, y_train)
score = metrics.accuracy_score(y_test, classifier.predict(X_test))
print('Accuracy: {0:f}'.format(score))
```

```python
f0_min = np.min(X_train[:,0])
f0_max = np.max(X_train[:,0])

f1_min = np.min(X_train[:,1])
f1_max = np.max(X_train[:,1])
```

```python
xx,yy = np.meshgrid(np.linspace(f0_min,f0_max,num=100),np.linspace(f1_min,f1_max,num=100))
xx = np.hstack((xx.reshape(100**2,1),yy.reshape(100**2,1)))
```

```python
xx.shape
```

```python
pred_space = classifier.predict_proba(xx)[:,1]
```

```python
xx,yy = np.meshgrid(np.linspace(f0_min,f0_max,num=100),np.linspace(f1_min,f1_max,num=100))
im = plt.contourf(xx,yy,pred_space.reshape(100,100),cmap='coolwarm',alpha=.5)
plt.scatter(xd[:,0],xd[:,1],c=y_data,cmap='coolwarm');
plt.colorbar(im)
```
