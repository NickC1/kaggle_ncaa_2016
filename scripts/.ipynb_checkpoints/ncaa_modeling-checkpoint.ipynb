# Train a model
We are going to sample this many times to get a good idea of how we will do in the competition.

```python
scores = []
logloss = []
adjlog = []
for ii in range(10000):
    pr,sc,prob,lg,adj = md.logistic_regression(F_merged[f_names],T_merged.values.ravel())
    scores.append(sc)
    logloss.append(lg)
    adjlog.append(adj)
```

```python
sc_mean = np.around(np.mean(scores),decimals=3)
sc_std = np.around(np.std(scores),decimals=3)
sc_text1 = '$\mu$ = ' + str(sc_mean)
sc_text2 = '$\sigma$ = ' + str(sc_std)

ll_mean = np.around(np.mean(logloss),decimals=3)
ll_std = np.around(np.std(logloss),decimals=3)
ll_text1 = '$\mu$ = ' + str(ll_mean)
ll_text2 = '$\sigma$ = ' + str(sc_std)

adj_ll_mean = np.around(np.mean(adjlog),decimals=3)
adj_ll_std = np.around(np.std(adjlog),decimals=3)

adj_text1 = '$\mu$ = ' + str(adj_ll_mean)
adj_text2 = '$\sigma$ = ' + str(adj_ll_std)
```

```python
fig,ax = plt.subplots()
ax.hist(scores,bins=25,alpha=.8);
fig.text(.2,.8,str(sc_text1))
fig.text(.2,.7,str(sc_text2))
plt.xlabel('Percent Correct')
plt.ylabel('Count')
#plt.savefig('../figures/score_distribution.png',bbox_inches='tight')
```

```python
plt.hist(logloss,color='r',bins=25,alpha=.5,label='Log Loss');
plt.hist(adjlog,color='b',bins=25,alpha=.5,label='Adjusted Log Loss');
plt.legend()
plt.xlabel('Log Loss')
plt.ylabel('Count')

plt.text(.61,800,'Log Loss')
plt.text(.61,700,str(ll_text1))
plt.text(.61,600,str(ll_text2))

plt.text(.61,450,'Adj Log Loss')
plt.text(.61,350,str(adj_text1))
plt.text(.61,250,str(adj_text2))

#plt.savefig('../figures/log_loss_distribution.png',bbox_inches='tight')
```

# Round 1 Evaluation
If the submission is for round 1 we can test it below.
The class StageOneScore was coded by Jeff Brown.

```python
import numpy as np
import pandas as pd

class StageOneScore:
    
    '''
    Files SampleSubmission.csv and TourneyCompactResults.csv should be present
    '''
    def __init__(self,subfile,resultfile):
        self.sample = pd.read_csv(subfile,index_col=False)
        self.results = pd.read_csv(resultfile,index_col=False)
        
        del self.results['Daynum']
        del self.results['Wscore']
        del self.results['Lscore']
        del self.results['Wloc']
        del self.results['Numot']
        
        self.grouped = self.results.groupby(['Season','Wteam']).aggregate(lambda x: tuple(x))
            
    def splitsmpl(self, smpl):
        splt = pd.DataFrame(list(smpl['Id'].str.split('_')),columns=['season','team1','team2'],dtype=np.int)
        splt['pred'] = smpl['Pred']
        return splt
    
    def beat(self,season,winner,loser):
        try:
            return loser in self.grouped.loc[int(season),int(winner)][0]
        except KeyError:
            return False
    
    def whichwon(self,season,team1,team2):
        if self.beat(season,team1,team2):
            return 1
        if self.beat(season,team2,team1):
            return 2
        return 0
    
    def score(self,pred,won):
        if won == 0:
            return 0
        if won == 1:
            return np.log(pred)
        return np.log(1.0-pred)
    
    def run(self):
        splitit = self.splitsmpl(self.sample)
        print 'splitit length = ',len(splitit.index)
        splitit['w'] = splitit.apply(lambda row: self.whichwon(row['season'],row['team1'],row['team2']),axis=1)
        print 'splitit w sum = ',splitit['w'].astype(bool).sum()
        splitit['score'] = splitit.apply(lambda row: self.score(row['pred'],row['w']),axis=1)
        a = -1.0 * splitit['score'].sum()
        print 'a = ',a
        b = splitit['score'].astype(bool).sum()
        print 'b = ',b
        return a/b
```

```python

s = StageOneScore('submission_one.csv','../data/TourneyCompactResults.csv')
#s.run()
```
